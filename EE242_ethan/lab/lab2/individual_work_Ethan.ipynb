```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage
from skimage import color
from skimage import io
from PIL import Image
```

```python
def SED(image):
    h1=  [[1,0,-1],[2,0,-2],[1,0,-1]]
    h2 = [[1,2,1],[0,0,0],[-1,-2,-1]]
    verti = ndimage.filters.convolve(image, h2)
    horiz = ndimage.filters.convolve(image, h1)
    ynm = np.sqrt(np.power(verti, 2)+np.power(horiz,2))
    return h1, h2, ynm
```

```python
img1 = color.rgb2gray(io.imread('img1.jpg'))

img2 = color.rgb2gray(io.imread('img2.jpg'))
```

```python
w, w, img1_e = SED(img1)
w, w, img2_e = SED(img2)
```

```python
fig = plt.figure(1, figsize=(10,10))

plt.subplot(221) 
plt.imshow(img1_e, cmap = 'gray')

plt.subplot(222) 
plt.imshow(img1, cmap = 'gray')

plt.subplot(223) 
plt.imshow(img2_e, cmap = 'gray')

plt.subplot(224) 
plt.imshow(img2, cmap = 'gray')

```

The image_1 has no edge. However, in the edge detector, the result shows a large amount of edge. I think this is because the image does not have a large color defference, and each slight increment in pixel would be meaningful to the detector. Therefore, the edge detector detect a large number of change.

The image_2 has edges. And the edge detector is able to detect the edge.
