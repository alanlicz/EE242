```python
import simpleaudio as sa
from scipy.io import wavfile as wav
import numpy as np
import cmath
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# individual work
fs, horn_mag = wav.read('tuba11.wav')
horn_mag = horn_mag[:,0]
t = np.arange(0,len(horn_mag)/fs, 1/fs)
print(horn_mag)

start_ind = round(1.3*fs)
end_ind = round(1.35*fs)

print(end_ind-start_ind)

x_1 = horn_mag[start_ind:start_ind+512]
t_1 = t[start_ind:start_ind+512]

print(len(x_1))

x_1_xhf = np.fft.fft(x_1)
x_1_xhf2 = np.fft.fftshift(x_1_xhf)
x_1_xhf2 = np.abs(x_1_xhf2)
x_1_positive = x_1_xhf2[256:]

w_x = np.arange(0,256,1)*fs/512

fig = plt.figure(1, figsize=(10,7))

# plot parameter set 1
plt.subplot(211) 
plt.plot(t_1, x_1)
plt.title('tuba11 sound sigment1, time region')
plt.xlabel('time(s)')
plt.ylabel('magnitude')

# plot parameter set 2
plt.subplot(212) 
plt.plot(w_x, x_1_positive)
plt.title('tuba11 sound sigment1, freq response')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()
```

```python
start_ind = round(7*fs)

x_2 = horn_mag[start_ind:start_ind+512]
t_2 = t[start_ind:start_ind+512]

x_2_xhf = np.fft.fft(x_2)
x_2_xhf2 = np.fft.fftshift(x_2_xhf)
x_2_xhf2 = np.abs(x_2_xhf2)
x_2_positive = x_2_xhf2[256:]

w_x2 = np.arange(0,256,1)*fs/512

fig = plt.figure(1, figsize=(10,7))

# plot parameter set 1
plt.subplot(211) 
plt.plot(t_2, x_2)
plt.title('tuba11 sound sigment2, time region')
plt.xlabel('time(s)')
plt.ylabel('magnitude')

# plot parameter set 2
plt.subplot(212) 
plt.plot(w_x2, x_2_positive)
plt.title('tuba11 sound sigment2, freq response')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()
```
