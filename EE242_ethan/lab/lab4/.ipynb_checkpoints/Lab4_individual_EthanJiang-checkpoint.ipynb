```python
from scipy.io import wavfile as wav
import numpy as np
import matplotlib.pyplot as plt
import simpleaudio as sa
import decimal
import math
from scipy import signal as sig
%matplotlib inline
```

```python
# Part A

W1 = 0.25
W3 = [0.25, 0.35]
W0 = [0.35, 0.45]
W4 = [0.45, 0.55]
W2 = 0.55

b_l,a_l = sig.butter(12, W1, 'lowpass')
b_0,a_0 = sig.butter(12, W0, 'bandpass')
b_b,a_b = sig.butter(12, W3, 'bandpass')
b_4,a_4 = sig.butter(12, W4, 'bandpass')
b_h,a_h = sig.butter(12, W2, 'highpass')


w_l, hf_l = sig.freqz(b_l, a_l)
w_h, hf_h = sig.freqz(b_h, a_h)
w_0, hf_0 = sig.freqz(b_0, a_0)
w_b, hf_b = sig.freqz(b_b, a_b)
w_4, hf_4 = sig.freqz(b_4, a_4)
w_b = w_b[1:]
hf_b = hf_b[1:]
w_h = w_h[1:]
hf_h = hf_h[1:]
w_0 = w_0[1:]
hf_0 = hf_0[1:]


plt.plot(w_l, 20 * np.log10(np.abs(hf_l)), label='low pass')
plt.plot(w_h, 20 * np.log10(np.abs(hf_h)), label='band pass 1')
plt.plot(w_0, 20 * np.log10(np.abs(hf_0)), label='band pass 2')
plt.plot(w_4, 20 * np.log10(np.abs(hf_4)), label='band pass 3')
plt.plot(w_b, 20 * np.log10(np.abs(hf_b)), label='band pass')



plt.title('HdB vs freq')
plt.ylim(-30, 2)
plt.xlabel('freq,radians')
plt.ylabel('HdB')
plt.legend() 
    
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()
```
