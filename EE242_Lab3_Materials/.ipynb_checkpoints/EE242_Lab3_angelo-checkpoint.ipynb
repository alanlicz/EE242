## Assignment 1

```python
import math as math
# a "game plan" suggestion: 

#step 1: 
a_triangle=np.zeros(21,complex)
a_sawtooth=np.zeros(21,complex)
#these are two vectors of 21 complex numbers. fill them out with the first 20 harmonics' complex amplitudes + DC offset 
#of the triangle and sawtooth waves from lab manual 
#these values given directly from provided mathematical expression formula
#suggestion: use for-loop using complex number operations 
#alternatively one could tediously enter them in manually individually and laboriously.

a_sawtooth[0] = 0.5
#a_sawtooth.append(0.5)

for k in range(1, 21):
    numerator = 1
    denominator = 1j * 2 * k * math.pi
    num = numerator/denominator
    a_sawtooth[k] = num

a_triangle[0] = 0.5
#a_triangle.append(0.5)

for k in range(1, 21):
    numerator = 2*math.sin(k * math.pi/2)
    denominator = 1j * (k * math.pi)**2
    coefficient = math.e ** ((-1j * 2 * k * math.pi)/2)
    num = coefficient * numerator/denominator
    a_triangle[k] = num
        

#step 2:
#we want 200 milliseconds of audio sound at sampling rate 8000 samples per second. 
#compute how many total samples are needed. (13 is just a placeholder)
totalsamples= int(.2 * 8000)
t=np.arange(totalsamples)*(1/8000)
x=np.zeros(totalsamples)
#x is the real output, right now it's a bunch of zeros, we need to fill it by adding the desired cosine waves to it.  
#write a for loop to add the desired number of harmonics or cosine waves to the x output 
#directly using the phases and magnitudes computed previously in a_triangle and a_sawtooth 
#to facilitate this process, note that np.cos(t) will take the cosine of each t value element wise in the t vector.  

x_t_N2 = np.array([])
value = 0.5
for n in range(1, 3):
    value += (2 * np.abs(a_triangle[n]) * np.cos(n * (2 * np.pi * 20 * t) + np.angle(a_triangle[n])))
    
x_t_N2 = value

x_t_N5 = np.array([])
value = 0.5
for n in range(1, 6):
    value += (2 * np.abs(a_triangle[n]) * np.cos(n * (2 * np.pi * 20 * t) + np.angle(a_triangle[n])))
    
x_t_N5 = value

x_t_N20 = np.array([])
value = 0.5
for n in range(1, 6):
    value += (2 * np.abs(a_triangle[n]) * np.cos(n * (2 * np.pi * 20 * t) + np.angle(a_triangle[n])))
    
x_t_N20 = value

x_s_N2 = np.array([])
value = 0.5
for n in range(1, 3):
    value += (2 * np.abs(a_sawtooth[n]) * np.cos(n * (2 * np.pi * 20 * t) + np.angle(a_sawtooth[n])))
    
x_s_N2 = value

x_s_N5 = np.array([])
value = 0.5
for n in range(1, 6):
    value += (2 * np.abs(a_sawtooth[n]) * np.cos(n * (2 * np.pi * 20 * t) + np.angle(a_sawtooth[n])))
    
x_s_N5 = value

x_s_N20 = np.array([])
value = 0.5
for n in range(1, 21):
    value += (2 * np.abs(a_sawtooth[n]) * np.cos(n * (2 * np.pi * 20 * t) + np.angle(a_sawtooth[n])))
    
x_s_N20 = value

#optionally once step 2 is working one can turn those tested-functional lines of code into a function to call up later. 
#remember what a function does: you feed in the data or input (e.g. a_triangle), and then: 
#it will hopefully spit out the corresponding desired output: (e.g. x and/or t) 

#this is how you plot:
fig = plt.figure(6, figsize = (10,5))

plt.figure(figsize = (15, 15))
plt.subplot(711)
plt.plot(t,x)
plt.title('x(t)')

plt.figure(figsize = (15, 15))
plt.subplot(712)
plt.plot(t, x_t_N2)
plt.title('Triangle N = 2')

plt.figure(figsize = (15, 15))
plt.subplot(713)
plt.plot(t, x_t_N5)
plt.title('Triangle N = 5')

plt.figure(figsize = (15, 15))
plt.subplot(714)
plt.plot(t, x_t_N20)
plt.title('Triangle N = 20')

plt.figure(figsize = (15, 15))
plt.subplot(715)
plt.plot(t, x_s_N2)
plt.title('Sawtooth N = 2')

plt.figure(figsize = (15, 15))
plt.subplot(716)
plt.plot(t, x_s_N5)
plt.title('Sawtooth N = 5')

plt.figure(figsize = (15, 15))
plt.subplot(717)
plt.plot(t, x_s_N20)
plt.title('Sawtooth N = 20')
```

## Assignment 2

```python
def fun(t, a_k, f_0):  #f_0 fundamental freq
    x = a_k[0]
    w = 2*np.pi*f_0
    
    for k in range (1, len(a_k)):
        x = x + 2 * np.abs(a_k[k])*np.cos(k*w*t+np.angle(a_k[k]))
    return x

#a "game plan" suggestion:
#to save you all time, here are the coefficients from the lab manual:
ak_abs = [2688, 1900, 316, 178, 78, 38]
ak_phase =  [-1.73, -1.45, 2.36, 2.30, -2.30, 1.13]
ak_abs = np.array(ak_abs)
ak_phase = np.array(ak_phase)
ak = np.ones(len(ak_abs), dtype = complex) #3
ak = ak_abs * np.exp(1j*ak_phase)
ak = np.concatenate([[0],ak])
#again, we are adding a linear combination or superposition of cosine waves
#exactly the same as before and we even now have the coefficient list.
#analogous to previous: 
#100 milliseconds of total sound audio at 11025 samples per second corresponds to how many total samples?
#(again 13 is just a placeholder)
totalsamples=13
t=np.arange(13)*(1/11025)
x=np.zeros(13)

#Same as before: 
#write a for loop to add up the desired harmonics at which corresonding frequencies, magnitudes, and phases to place into the output x. 
#again, np.cos(t) will take the element-wise cosine value of each number in vector t, which should hopefully facilitate this process. 
#if this helps, the synthesis mathematical equation comes directly from assignment 1 to reuse here for assignment 2. 
fs, horn_mag = wav.read('horn11short.wav')

starting_index = round(0.2*fs)
ending_index = round(0.3*fs)
x1 = horn_mag[starting_index:ending_index + 1]

T = 0.1   #100ms
t = np.arange(0, T+1/fs, 1/fs) # new time vector of length 100ms
t = t + 0.2  # t starts at 0.2s, end at 0.3s
f0 = 335
w0 = 2*np.pi*f0

y = lambda t: np.sin(w0*t)
x = y(t)* 32767 # # NOTE: Multiply your concatenated notes (with amplitude of 1) with 32767 followed by conversion to int16 format
wav.write('sine.wav', fs, x.copy().astype('int16'))

sine_obj = sa.WaveObject.from_wave_file('sine.wav')     # Load audio file
fs1 = sine_obj.sample_rate                                  # Extract sampling frequency
channels = sine_obj.num_channels

x2 = fun(t, ak, f0).real
print(x2)
print(len(x2))

print('Sampling rate of sine: ' + str(fs1) + 'Hz')
sine_obj.play()

#this is how you plot:
fig = plt.figure(3, figsize=(10,5))
# plot parameter set 1
plt.subplot(211) 
plt.plot(t, x1)
plt.title('original sound signal')
plt.xlabel('Time(s)')
plt.ylabel('magnitude')


# plot parameter set 2
plt.subplot(212) 
plt.plot(t, x2)
plt.title('synthesized sound signal')
plt.xlabel('Time(s)')
plt.ylabel('magnitude')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()


```

```python
#a suggested "game plan":
#first step: from hornsound, take the 1024-length snippet of 2000 to 3024th samples
#(this roughly corresponds to the 200ms to 300ms chunk of the recording at sampling rate 11025) 
x3 = x1[:1024]

print(x3)
print(len(x3))
#x here below is just a temporary placheolder

xhf = np.fft.fft(x3)
xhf2 = np.fft.fftshift(xhf)
po_ne = np.abs(xhf2)
positive = po_ne[512:]
print(len(positive))

w = np.arange(-512,512,1)
print(len(w))
print(w)
print(w[512])

fig = plt.figure(4, figsize=(10,5))
# plot parameter set 1
plt.subplot(211) 
plt.plot(w* 11025/1024, po_ne)
plt.title('positive and negtive')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')


# plot parameter set 2
plt.subplot(212) 
plt.plot(w[512:]* 11025/1024, positive)
plt.title('positive')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()

#TL;DR fft transforms/maps the x 1024 time vector to 1024 frequency vector using 1024-by-1024 square matrix multiplication on vector
#the output is complex, the magnitude of which is symmetric (mirror imaged across the middle 512 mark)
#TL;DR fftshift basically takes frequency samples 1-thru-512 and shoves them into 513-1024 and vice versa 
#(e.g. the two separate 512 half-chunks swap places)

#we want to have a nice display:
#since these are complex vectors, display the magnitude using np.abs()

#two tasks: display the symmetric 1024-length spectrum

#we want to also display one-sided 512-length half spectrum (since symmetric just throw out second half redundancy)

#vertical axis display is clearly the magnitude

#horizontal axis display is described by frequency scaling method in the lab manual
#alternatively, horizontal axis labels can be done manually by using np.linspace
#just ensure commensurate bounds using intuition from Nyquist-Shannon sampling rate theory

#naturally use plt.plot(horizontal points, vertical points) for display, as demonstrated previously


#repeat same above procedure, but instead for a longer 2048 sample-snippet of hornsound, compare results
fig = plt.figure(5, figsize=(10,5))

plt.subplot(211) 
plt.plot(w[512:]* 11025/1024, positive)
plt.title('positive, log x-axis')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')
plt.xlim(200)

plt.subplot(212) 
plt.plot(w[512:]* 11025/1024, positive)
plt.title('positive')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()

x4 = horn_mag[starting_index:starting_index + 2048]

xhf_1 = np.fft.fft(x4)
xhf2_1 = np.fft.fftshift(xhf_1)
po_ne_1 = np.abs(xhf2_1)
positive_1 = po_ne_1[1024:]


w_1 = np.arange(-1024,1024,1)

fig = plt.figure(6, figsize=(20,5))

# plot parameter set 1
plt.subplot(211) 
plt.plot(w[512:]* 11025/1024, positive)
plt.title('positive, 1024, log x-axis')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')
plt.xlim(200)

# plot parameter set 2
plt.subplot(212) 
plt.plot(w_1[1024:] * 11025/2048, positive_1)
plt.title('positive, 2048, log x-axis')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')
plt.xlim(200)


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()
```

```python
#a suggested "game plan": 
#pick time indices of 0.75-0.78 seconds of audio at sampling rate 16000 
#take out this snippet which is the "oo" in "blue" or the land of oo in adventure time
#pick out time indices of 2.565-2.595 seconds of audio at sampling rate 16000
#this is the sound snippet that corresponds to the "s" in "nose"
#we want to frequency analyze these snippets and compare. 
#to do this, we use the fft function from before.
#remember fft works best with a nice power of 2: 512, 1024, 2048 examples from directly previous
#plot the spectra and the time series and compare, using same procedure as assignment 3 previously

fs, bluenose3_mag = wav.read('bluenose3.wav')
print(fs)
print(len(bluenose3_mag))
t = np.arange(0,len(bluenose3_mag)/fs, 1/fs)

fig = plt.figure(7, figsize=(10,5))

# plot parameter set 1
plt.subplot(111) 
plt.plot(t, bluenose3_mag)
plt.title('full bluenose3.wav')
plt.xlabel('time(s)')
plt.ylabel('magnitude')

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()

bluenose3 = sa.WaveObject.from_wave_file('bluenose3.wav')
bluenose3.play()

start_ind = round(0.75*16000)
print(start_ind)

sig_oo = bluenose3_mag[start_ind:start_ind+512]
t_oo = t[start_ind:start_ind+512]

sig_oo_xhf = np.fft.fft(sig_oo)
sig_oo_xhf2 = np.fft.fftshift(sig_oo_xhf)
sig_oo_xhf2 = np.abs(sig_oo_xhf2)
sig_oo_positive = sig_oo_xhf2[256:]

w_oo = np.arange(0,256,1)*fs/512

fig = plt.figure(8, figsize=(10,7))

# plot parameter set 1
plt.subplot(211) 
plt.plot(t_oo, sig_oo)
plt.title('oo sound, time region')
plt.xlabel('time(s)')
plt.ylabel('magnitude')

# plot parameter set 2
plt.subplot(212) 
plt.plot(w_oo, sig_oo_positive)
plt.title('oo sound, freq response')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()

start_ind = round(2.565*16000)
print(start_ind)

sig_s = bluenose3_mag[start_ind:start_ind+512]
t_s = t[start_ind:start_ind+512]

sig_s_xhf = np.fft.fft(sig_s)
sig_s_xhf2 = np.fft.fftshift(sig_s_xhf)
sig_s_xhf2 = np.abs(sig_s_xhf2)
sig_s_positive = sig_s_xhf2[256:]

w_s = np.arange(0,256,1)*fs/512

fig = plt.figure(9, figsize=(10,7))

# plot parameter set 1
plt.subplot(211) 
plt.plot(t_s, sig_s)
plt.title('s sound, time region')
plt.xlabel('time(s)')
plt.ylabel('magnitude')

# plot parameter set 2
plt.subplot(212) 
plt.plot(w_s, sig_s_positive)
plt.title('s sound, freq response')
plt.xlabel('freq(Hz)')
plt.ylabel('magnitude')
plt.xscale('log')


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=2)
plt.show()
```
